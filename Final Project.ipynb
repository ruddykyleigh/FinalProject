```python
## IMPORTS ##
import numpy as np
import matplotlib.pyplot as plt

##LOAD DATA##
file = open("CA_TREES (1).txt", 'r') #read the file
next(file) #skip the headers
sapling_data = [] #create arrays for both types of data to separate by variable
saplingless_data = []

### READ THE FILE ###
for line in file: #loop
    values = line.strip().split('\t') #split values by tab
    relevant_data = [float(values[9]), float(values[10]), float(values[12])] #extract the three variables 
    if values[13] == "Sapling": #check the status column
        sapling_data.append(relevant_data) #put everything into the associated list
    elif values[13] == "Saplingless":
        saplingless_data.append(relevant_data)

### CENTRAL TENDENCY PER STATUS ###
sapling_data = np.array(sapling_data) #make the lists into numpy arrays
saplingless_data = np.array(saplingless_data)

#calculate means
sapling_mean = np.mean(sapling_data, axis=0) 
saplingless_mean = np.mean(saplingless_data, axis=0) 

#calculate median
sapling_median = np.median(sapling_data, axis=0) 
saplingless_median = np.median(saplingless_data, axis=0) 

#calculate standard deviation
sapling_std_dev = np.std(sapling_data, axis=0) 
saplingless_std_dev = np.std(saplingless_data, axis=0) 

print(f"Sapling Means (Slope factor (%), TWI, and Solar Area (WH/m2)): {sapling_mean}")
print(f"Saplingless Means (Slope factor (%), TWI, and Solar Area (WH/m2)): {saplingless_mean}")
print(f"Sapling Median (Slope factor (%), TWI, and Solar Area (WH/m2)): {sapling_median}")
print(f"Saplingless Median (Slope factor (%), TWI, and Solar Area (WH/m2)): {saplingless_median}")
print(f"Sapling Standard Deviation (Slope factor (%), TWI, and Solar Area (WH/m2)): {sapling_std_dev}")
print(f"Saplingless Standard Deviation (Slope factor (%), TWI, and Solar Area (WH/m2)): {saplingless_std_dev}")

#### HISTOGRAM ####
file = open("CA_TREES (1).txt", 'r') #read the file
x=int(input("Input variable index:")) #asks user for the variable 
y=str(input("Input Sapling or Saplingless:")) #asks user for which status they need

if y == "Sapling": #defines names for each of the columns
    if x == 9:
        name = "Sapling Slope Factor (%)"
    elif x == 10:
        name = "Sapling TWI"
    elif x == 12:
        name = " Sapling Solar Area (WH/m2)"
elif y == "Saplingless":
    if x == 9:
        name = "Saplingless Slope Factor (%)"
    elif x == 10:
        name = "Saplingless TWI"
    elif x == 12:
        name = "Saplingless Solar Area (WH/m2)"
    
next(file) #skip the headers
results = [] #creates a list for the results
line_counter = 0 #initializes the counter

if y == "Saplingless": 
    for line in file: #loop
        if line_counter <= 33: #only takes saplingless results, from lines 1-34
            values = line.strip().split('\t') #split values by tab         
            results.append(float(values[x])) #adds values to list
            line_counter = line_counter + 1 #continues counter
        elif line_counter > 32:
            break        
elif y == "Sapling":
    for line in file:
        if line_counter <= 32: #only takes sapling results, from lines 35-90
            results = []
            line_counter = line_counter + 1
        elif 33 <= line_counter <= 89:
            values = line.strip().split('\t') #split values by tab         
            results.append(float(values[x]))
            line_counter = line_counter + 1
        elif line_counter > 89:
            break
        
plt.hist(results, bins="auto", density=False)  #generates a historam, from the data (results) with automatic bins arguments are passed to np.histogram
title="Histogram of " + name + " with 'auto' bins"   #variable name for the title of the graph
plt.title(title)    #adds title to the graph
plt.show() #not necessary for Jupyter
        
##CORRELATION##
import numpy as np
from scipy import stats
file = open("CA_TREES (1).txt", 'r') #read the file
x=int(input("Input first variable index:")) #asks user for the variables
y=int(input("Input second variable index:"))
z=str(input("Input Sapling or Saplingless:")) #asks user for which status they need #asks user for which status they need

next(file) #skip the headers
results_x = [] #creates a list for the results
results_y = []
line_counter = 0 #initializes the counter

if z == "Saplingless": 
    for line in file: #loop
        if line_counter <= 32: #only takes saplingless results, from lines 1-34
            values = line.strip().split('\t') #split values by tab         
            results_x.append(float(values[x])) #adds values to list
            results_y.append(float(values[y])) #adds values to list
            line_counter = line_counter + 1 #continues counter
        elif line_counter > 32:
            break        
elif z == "Sapling":
    for line in file:
        if line_counter <= 32: #only takes sapling results, from lines 35-90
            results = []
            line_counter = line_counter + 1
        elif 33 <= line_counter <= 89:
            values = line.strip().split('\t') #split values by tab         
            results_x.append(float(values[x])) #adds values to list
            results_y.append(float(values[y])) #adds values to list
            line_counter = line_counter + 1
        elif line_counter > 89:
            break

##PEARSON##
res = stats.pearsonr(results_x, results_y) #pearson correlation
print("The pearson correlation for variables in indices", x, "and", y, "for status", z, "is", res) #prints results

#SPEARMAN##
results_x = np.array(results_x) #put the lists into numpy arrays
results_y = np.array(results_y)
res = stats.spearmanr(results_x, results_y) #show results for spearmans
print("The spearman correlation for variables in indices", x, "and", y, "for status", z, "is", res)

##LINEAR REGRESSION##
linregress = stats.linregress(results_x, results_y)

slope, intercept, r_value, p_value, std_err = linregress

# Output results
print("The following results are from the linear regression output:")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")

##SCATTERPLOT##
import matplotlib.pyplot as plt
import numpy as np
file = open("CA_TREES (1).txt", 'r') #read the file
x=int(input("Input first variable index:")) #asks user for the variables
z=int(input("Input second variable index:")) #asks user for the variables
y=str(input("Input Sapling or Saplingless:")) #asks user for which status they need #asks user for which status they need

if y == "Sapling": #defines names for each of the columns
    if x == 9:
        y_axis = "Sapling Slope Factor (%)"
    elif x == 10:
        y_axis = "Sapling TWI"
    elif x == 12:
        y_axis = " Sapling Solar Area (WH/m2)"
        
    if z == 9:
        x_axis = "Sapling Slope Factor (%)"
    elif z == 10:
        x_axis = "Sapling TWI"
    elif z == 12:
        x_axis = "Sapling Solar Area (WH/m2)"
elif y == "Saplingless":
    if x == 9:
        y_axis = "Saplingless Slope Factor (%)"
    elif x == 10:
        y_axis = "Saplingless TWI"
    elif x == 12:
        y_axis = "Saplingless Solar Area (WH/m2)"
        
    if z == 9:
        x_axis = "Saplingless Slope Factor (%)"
    elif z == 10:
        x_axis = "Saplingless TWI"
    elif z == 12:
        x_axis = " Saplingless Solar Area (WH/m2)"

name = y_axis + " vs " + x_axis        
next(file) #skip headers
results=[]         # This will hold the measurements for the column of interest
region=[]         # holds measurements for x-axis
line_counter = 0

if y == "Saplingless": 
    for line in file: #loop
        if line_counter <= 32: #only takes saplingless results, from lines 1-34
            values = line.strip().split('\t') #split values by tab         
            results.append(float(values[x])) #adds values to list
            region.append(float(values[z]))
            line_counter = line_counter + 1 #continues counter
        elif line_counter > 32:
            break        
elif y == "Sapling":
    for line in file:
        if line_counter <= 32: #only takes sapling results, from lines 35-90
            results = []
            line_counter = line_counter + 1
        elif 33 <= line_counter <= 89:
            values = line.strip().split('\t') #split values by tab         
            results.append(float(values[x]))
            region.append(float(values[z]))
            line_counter = line_counter + 1
        elif line_counter > 89:
            break

        
results = np.array(results)
region = np.array(region)

### GENERATE PLOT ###
fig, ax = plt.subplots()  # Create a figure containing a single Axes.
ax.scatter(region,results)  # Plot some data on the Axes.
m, b = np.polyfit(region, results, 1)
# Plot the regression line
ax.plot(region, m*region + b, color='red')
ax.set_xlabel(x_axis) #set label names
ax.set_ylabel(y_axis)
plt.title(name)
plt.show()  #show figure

##Run PCA##
import numpy as np
from sklearn.decomposition import PCA
file = open("CA_TREES (1).txt", 'r') #read the file
next(file)
line_counter = 0
saplingless_results = []
sapling_results = []


for line in file: #loop
    if line_counter <= 32: #only takes saplingless results, from lines 1-34
        values = line.strip().split('\t') #split values by tab         
        relevant_data=[float(values[9]), float(values[10]), float(values[12])] #adds values to list
        saplingless_results.append(relevant_data)
        line_counter = line_counter + 1 #continues counter
    elif line_counter > 32:
        break        

for line in file:
    if line_counter <= 32: #only takes sapling results, from lines 35-90
        sapling_results = []
        line_counter = line_counter + 1
    elif 33 <= line_counter <= 89:
        values = line.strip().split('\t') #split values by tab         
        relevant_data=[float(values[9]), float(values[10]), float(values[12])] #adds values to list
        sapling_results.append(relevant_data)
        line_counter = line_counter + 1
    elif line_counter > 89:
        break

saplingless_results = np.array(saplingless_results)
sapling_results = np.array(sapling_results)


# Initialize PCA object
pca = PCA(n_components=3)   ### Change or remove this to get more components
pca.fit(saplingless_results)
principal_components_saplingless = pca.fit_transform(saplingless_results)
print(principal_components_saplingless)
print("Saplingless PCA is described by ",pca.explained_variance_ratio_)
plt.hist(pca.explained_variance_ratio_, bins="auto", density=False)  # Generates a historam, from the data (results) with automatic bins arguments 
title="Bar Chart Showing PCA Weights for Saplingless Data"  
plt.title(title)   

plt.show()    # Not required for Jupyter

pca = PCA(n_components=3)
pca.fit(sapling_results)
principal_components_sapling = pca.fit_transform(sapling_results)
print(principal_components_sapling)
print("Sapling PCA is described by ",pca.explained_variance_ratio_)
plt.hist(pca.explained_variance_ratio_, bins="auto", density=False)  # Generates a historam, from the data (results) with automatic bins arguments 
title="Bar Chart Showing PCA Weights for Sapling Data"
plt.title(title)    

### Plot PCA ###
### IMPORTS ###
import matplotlib.pyplot as plt

### GENERATE PLOT ###
fig, ax = plt.subplots()             # Create a figure containing a single Axes.

x=str(input("Input status of desired PCA: "))
y=int(input("Enter first variable to display: "))
z=int(input("Enter second variable to display: "))

if y == 9:
    PCX=0
    xname = "Slope Factor"
elif y == 10:
    PCX = 1
    xname = "TWI"
elif y == 12:
    PCX = 2
    xname = "Solar Area"

if z == 9:
    PCY=0
    yname = "Slope Factor"
elif z == 10:
    PCY = 1
    yname = "TWI"
elif z == 12:
    PCY = 2
    yname = "Solar Area"

if x == "Saplingless":
    ax.scatter(principal_components_saplingless[:15 , PCX], principal_components_saplingless[:15 , PCY], c="blue")  # Plot some data on the Axes.
    ax.scatter(principal_components_saplingless[15:30 , PCX], principal_components_saplingless[15:30 , PCY], c="green")  # Plot some data on the Axes.
    ax.scatter(principal_components_saplingless[30: , PCX], principal_components_saplingless[30: , PCY], c="red")  # Plot some data on the Axes.

    ### Set Axis Labels ###
    label_X = "PCX " + xname + " Saplingless (" + str(np.round(pca.explained_variance_ratio_[PCX]*100, decimals=2)) + "%)"
    label_Y = "PCY " + yname + " Saplingless (" + str(np.round(pca.explained_variance_ratio_[PCY]*100, decimals=2)) + "%)"
    ax.set_xlabel(label_X)
    ax.set_ylabel(label_Y)
elif x == "Sapling":
    ax.scatter(principal_components_sapling[:26, PCX], principal_components_sapling[:26 , PCY], c="blue")  # Plot some data on the Axes.
    ax.scatter(principal_components_sapling[26: , PCX], principal_components_sapling[26: , PCY], c="green")  # Plot some data on the Axes.

    ### Set Axis Labels ###
    label_X = "PCX " + xname + " Sapling (" + str(np.round(pca.explained_variance_ratio_[PCX]*100, decimals=2)) + "%)"
    label_Y = "PCY " + yname + " Sapling (" + str(np.round(pca.explained_variance_ratio_[PCY]*100, decimals=2)) + "%)"
    ax.set_xlabel(label_X)
    ax.set_ylabel(label_Y)

```
